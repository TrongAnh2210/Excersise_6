```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
url = '/content/gdrive/MyDrive/colab/mtcars.csv'
data = pd.read_csv(url)
data.hist()
plt.show
correlation = data.corr() # Mối tương quan giữa các biến 
# Số dương tỉ lệ thuận , số âm tỉ lệ nghịch 
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['wt']), y = data['mpg'])
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
y_pred = model.predict(X = pd.DataFrame(data['wt']) )
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
model.fit(X,y=data['mpg'])
print('intercept của mô hình: ', model.intercept_)
print('coefient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y=data['mpg']))
```

```python
range = np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range,range**2,range**3]).T # T = chuyển vị
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black',xlim= (0,7))
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5]).T
model.fit(X,y=data['mpg'])
print('intercept của mô hình: ', model.intercept_)
print('coefient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y=data['mpg']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5, data['wt']**6,  data['wt']**7,  data['wt']**8,  data['wt']**9,
                   data['wt']**10]).T
model.fit(X,y=data['mpg'])
print('intercept của mô hình: ', model.intercept_)
print('coefient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y=data['mpg']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, data['wt']**4, data['wt']**5, data['wt']**6,  data['wt']**7,  data['wt']**8,  data['wt']**9,
                   data['wt']**10,  data['wt']**11,  data['wt']**12,data['wt']**13,data['wt']**14,data['wt']**15 ]).T
model.fit(X,y=data['mpg'])
print('intercept của mô hình: ', model.intercept_)
print('coefient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y=data['mpg']))
```

```python
model = linear_model.LinearRegression()
model.fit(X = data.loc[:, ['wt','hp','disp']], y = data['mpg'])
print(model.intercept_)
print(model.coef_)
print(model.score(X = data.loc[:, ['wt','hp','disp']], y = data['mpg']))
```

### SALARY_DATA


```python
url = '/content/gdrive/MyDrive/colab/Salary_Data.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame(data['Salary'])
y = data['YearsExperience']
model.fit(X, y )
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
```

```python
range = np.arange(20000,130000,1)
predictor = pd.DataFrame([range]).T # T = chuyển vị
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2]).T
y = data['YearsExperience']
model.fit(X, y )
print(model.intercept_)
print(model.coef_)
model.score(X, y)
```

```python
range = np.arange(20000,130000,1)
predictor = pd.DataFrame([range, range**2]).T # T = chuyển vị
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3]).T
y = data['YearsExperience']
model.fit(X, y )
print(model.intercept_)
print(model.coef_)
model.score(X, y)
```

```python
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
y_pred = model.predict(X).round(1)
caculation = pd.DataFrame(np.c_[y, y_pred], columns= ['Original Salary','Predicted Salary'])
caculation
```

```python
range = np.arange(20000,130000,1)
predictor = pd.DataFrame([range, range**2,range**3 ]).T # T = chuyển vị
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

### 50_STARTUPS_DATA



```python
url = '/content/gdrive/MyDrive/colab/50_Startups.csv'
data = pd.read_csv(url)
data.drop("State", axis = 1, inplace = True)
data.hist()
plt.show
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x = 'Profit', y = 'R&D Spend', figsize = (9,9), color = 'black')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'], data['Profit']**2, data['Profit']**3]).T
y = data['R&D Spend']
model.fit(X, y )
print(model.intercept_)
print(model.coef_)
model.score(X, y)
```

```python
range = np.arange(data['Profit'].min(),data['Profit'].max(),10)
predictor = pd.DataFrame([range, range**2, range**3]).T # T = chuyển vị
data.plot(kind = 'scatter', x = 'Profit', y = 'R&D Spend', figsize = (9,9), color = 'black')
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
model.fit(X = data.loc[:, ['R&D Spend','Administration','Marketing Spend']], y = data['Profit'])
print(model.intercept_)
print(model.coef_)
print(model.score(X = data.loc[:, ['R&D Spend','Administration','Marketing Spend']], y = data['Profit']))
```
